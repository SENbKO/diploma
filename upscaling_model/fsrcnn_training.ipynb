```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from tqdm import tqdm
from FSRCNN import FSRCNN
from dataset_loader import SRDataset
import matplotlib.pyplot as plt
import pickle  
from pathlib import Path

HR_DIR = Path(__file__).parent / ".." / "dataset" / "DIV2K_train_HR" / "DIV2K_train_HR"
SCALE = 2
CROP_SIZE = 96
BATCH_SIZE = 16
EPOCHS = 80
LR = 1e-4

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

dataset = SRDataset(HR_DIR, scale=SCALE, crop_size=CROP_SIZE)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)


model = FSRCNN(scale_factor=SCALE).to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=LR)

batch_losses = []  
epoch_losses = [] 

for epoch in range(EPOCHS):
    model.train()
    epoch_loss = 0

    loop = tqdm(loader, desc=f"Epoch [{epoch+1}/{EPOCHS}]")
    for lr_y, hr_y in loop:
        lr_y = lr_y.to(device)
        hr_y = hr_y.to(device)

        optimizer.zero_grad()
        sr_y = model(lr_y)

        loss = criterion(sr_y, hr_y)
        loss.backward()
        optimizer.step()

        batch_losses.append(loss.item())
        epoch_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    avg_epoch_loss = epoch_loss / len(loader)
    epoch_losses.append(avg_epoch_loss)
    print(f"Epoch {epoch+1}: Average Loss = {avg_epoch_loss:.6f}")

# Save model
torch.save(model.state_dict(), "fsrcnn_x2_y.pth")
print("Model saved as fsrcnn_x2_y.pth")

losses_dict = {
    "batch_losses": batch_losses,
    "epoch_losses": epoch_losses
}

with open("fsrcnn_training_losses.pkl", "wb") as f:
    pickle.dump(losses_dict, f)

print("Training losses saved in fsrcnn_training_losses.pkl")


plt.figure(figsize=(10,5))
plt.plot(batch_losses, label="Batch Loss", color="blue", alpha=0.7)
plt.plot([i*len(loader) for i in range(EPOCHS)], epoch_losses, label="Epoch Avg Loss", color="red", linewidth=2)
plt.xlabel("Batch Number")
plt.ylabel/("MSE Loss")
plt.title("FSRCNN ×2 Training Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

def average_batches(x, window=50):
    return np.convolve(x, np.ones(window)/window, mode="valid")

plt.figure(figsize=(12,5))
plt.plot(average_batches(batch_losses), label="Training Loss")
plt.xlabel("Training Step")
plt.ylabel("Training loss")
plt.title("FSRCNN training loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
import numpy as np
from PIL import Image
from skimage.metrics import peak_signal_noise_ratio as psnr
import matplotlib.pyplot as plt
from torchvision.transforms.functional import to_tensor, to_pil_image
from FSRCNN import FSRCNN
import time

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = FSRCNN(scale_factor=2).to(device)
model.load_state_dict(torch.load("fsrcnn_x2_y.pth", map_location=device))
model.eval()


def rgb2y_np(img_np):
    return 0.299 * img_np[..., 0] + 0.587 * img_np[..., 1] + 0.114 * img_np[..., 2]

def test_image(image_path, save_path="sr.png"):
    hr_img = Image.open(image_path).convert("YCbCr")
    w, h = hr_img.size

    w = (w // 2) * 2
    h = (h // 2) * 2
    hr_img = hr_img.crop((0, 0, w, h))

    y_hr, cb, cr = hr_img.split()

    y_lr = y_hr.resize((w // 2, h // 2), Image.BICUBIC)
    y_lr_tensor = to_tensor(y_lr).unsqueeze(0).to(device)

    if device.type == "cuda":
        torch.cuda.synchronize()
    start_time = time.perf_counter()

    with torch.no_grad():
        y_sr = model(y_lr_tensor).clamp(0, 1)

    if device.type == "cuda":
        torch.cuda.synchronize()
    end_time = time.perf_counter()

    upscale_time_ms = (end_time - start_time) * 1000

    y_sr_pil = to_pil_image(y_sr.squeeze(0).cpu()).convert("L")

    cb_up = cb.resize((w, h), Image.BICUBIC)
    cr_up = cr.resize((w, h), Image.BICUBIC)

    sr_img = Image.merge("YCbCr", (y_sr_pil, cb_up, cr_up)).convert("RGB")

    y_lr_up = y_lr.resize((w, h), Image.BICUBIC)
    lr_rgb = Image.merge("YCbCr", (y_lr_up, cb_up, cr_up)).convert("RGB")

    hr_rgb = hr_img.convert("RGB")

    lr_rgb.save("squirrel_lr.png")
    sr_img.save(save_path)
    hr_rgb.save("squirrel_hr.png")

    # PSNR
    sr_np = np.array(sr_img) / 255.0
    hr_np = np.array(hr_rgb) / 255.0

    print("Image sizes:")
    print(f"  LR: {y_lr.size[0]} × {y_lr.size[1]}")
    print(f"  SR: {sr_img.size[0]} × {sr_img.size[1]}")
    print(f"  HR: {hr_rgb.size[0]} × {hr_rgb.size[1]}")
    print()
    print(f"FSRCNN ×2 PSNR (RGB): {psnr(hr_np, sr_np):.2f} dB")
    print(f"FSRCNN ×2 PSNR (Y):   {psnr(rgb2y_np(hr_np), rgb2y_np(sr_np)):.2f} dB")
    print(f"Upscaling time: {upscale_time_ms:.2f} ms")

    plt.figure(figsize=(18, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(lr_rgb)
    plt.title(f"Downscaled by Factor 2 Original Image\n{y_lr.size[0]}×{y_lr.size[1]}")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(sr_img)
    plt.title(
        f"Upscaled by FSRCNN ×2 image\n{sr_img.size[0]}×{sr_img.size[1]} | {upscale_time_ms:.2f} ms"
    )
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(hr_rgb)
    plt.title(f"Original High Resolution image\n{hr_rgb.size[0]}×{hr_rgb.size[1]}")
    plt.axis("off")

    plt.show()

# Run test
test_image("0801.png")
```
